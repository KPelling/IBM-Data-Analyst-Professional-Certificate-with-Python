# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here

#Query to obtain ConvertedComp data from table.
QUERY = """
SELECT ConvertedComp
FROM master
"""

#Store data in data frame.
df = pd.read_sql_query(QUERY,conn)

#Import the necessities.
%matplotlib inline 
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

#Set style of plot.
mpl.style.use(['ggplot'])

#Replace missing values in ConvertedComp with the mean of the column.
df['ConvertedComp'].replace(np.nan, df.ConvertedComp.mean(), inplace = True)

#Reset index.
df.reset_index(drop=True, inplace=True)

#Initialize list of salaries and index.
comp_list = []
index = 0

#Looping through data frame...
for ConvertedComp in df.iterrows():
    
    #If salary falls into given range, assign respective group number to the list.
    if df['ConvertedComp'][index] >= 0 and df['ConvertedComp'][index] < 100000:
        comp_list.append(1)
        
    if df['ConvertedComp'][index] >= 100000 and df['ConvertedComp'][index] < 200000:
        comp_list.append(2)
        
    if df['ConvertedComp'][index] >= 200000 and df['ConvertedComp'][index] < 300000:
        comp_list.append(3)

    if df['ConvertedComp'][index] >= 300000 and df['ConvertedComp'][index] < 400000:
        comp_list.append(4)
        
    if df['ConvertedComp'][index] >= 400000 and df['ConvertedComp'][index] < 500000:
        comp_list.append(5)
        
    if df['ConvertedComp'][index] >= 500000 and df['ConvertedComp'][index] < 600000:
        comp_list.append(6)
        
    if df['ConvertedComp'][index] >= 600000 and df['ConvertedComp'][index] < 700000:
        comp_list.append(7)
        
    if df['ConvertedComp'][index] >= 700000 and df['ConvertedComp'][index] < 800000:
        comp_list.append(8)
    
    if df['ConvertedComp'][index] >= 800000 and df['ConvertedComp'][index] < 900000:
        comp_list.append(9)
        
    if df['ConvertedComp'][index] >= 900000:
        comp_list.append(10)
    
    #Increment index by one.
    index = index + 1

#Load list into new column to categorize each person in a salary group (1 being the lowest, 10 being the highest).
df['GroupedConvertedComp'] = comp_list

#Plot the graph using the grouped values.
df.GroupedConvertedComp.plot(kind='hist', stacked=False, figsize=(20, 10))

plt.title('Histogram of ConvertedComp')
plt.ylabel('Number of People')
plt.xlabel('Salary Group')

plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

#Query to obtain Age data from table.
QUERY = """
SELECT Age
FROM master
"""

#Store data in data frame.
df = pd.read_sql_query(QUERY,conn)

#Plot df.Age.
df.Age.plot(kind='box', figsize=(8, 6))

plt.title('Age Box Plot')
plt.ylabel('Age')

plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

#Query to obtain Age and WorkWeekHrs. data from table.
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""

#Store data in data frame.
df = pd.read_sql_query(QUERY,conn)

#Plot Age/WorkWeekHrs. scatter plot.
df.plot(kind='scatter', x='WorkWeekHrs', y='Age', figsize=(10, 6), color='darkblue')

plt.title('Age vs. Work Week Hours')
plt.xlabel('Work Week Hours')
plt.ylabel('Age')

plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

#Query to obtain CodeRevHrs, Age, and WorkWeekHrs data from table.
QUERY = """
SELECT CodeRevHrs, Age, WorkWeekHrs
FROM master
"""

#Store data in data frame.
df = pd.read_sql_query(QUERY,conn)

#Plot CodeRevHrs/WorkWeekHrs. scatter plot.
plt.scatter(df.CodeRevHrs, df.WorkWeekHrs, color='purple', alpha=.5, s = df.Age)

plt.ylabel("Code Rev Hours", size=14)
plt.xlabel("Work Week Hours", size=14)
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

#Query to obtain top 5 databases.
QUERY = """
SELECT DatabaseDesireNextYear, COUNT(Respondent) AS Total
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY COUNT(Respondent) DESC
LIMIT 5
"""

#Store data in data frame.
df = pd.read_sql_query(QUERY,conn)

#Create pie chart.
df['Total'].plot(kind='pie', figsize=(5, 6), autopct='%1.1f%%', startangle=90, shadow=True)

plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.axis('equal')
plt.legend(labels=df.DatabaseDesireNextYear, loc='upper left')

plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

#Query to obtain WorkWeekHrs and CodeRevHrs for people between 30 and 35.
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE Age >= 30 AND Age <=35
"""

#Store data in data frame.
df = pd.read_sql_query(QUERY,conn)

#Find medians of WorkWeekHrs and CodeRevHrs by age.
df.groupby('Age').median().plot.bar(stacked=True)
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

#Query to obtain ConvertedComp for people between 45 and 60.
QUERY = """
SELECT ConvertedComp, Age
FROM master
WHERE Age >= 45 AND Age <=60
"""

#Store data in data frame.
df = pd.read_sql_query(QUERY,conn)

#Find medians of ConvertedComp by age.
df.groupby('Age').median().plot.line(stacked=False)
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

#Query to obtain MainBranch data.
QUERY = """
SELECT MainBranch, COUNT(Respondent) AS NumRespondents
FROM master
GROUP BY MainBranch
"""

#Store data in data frame.
df = pd.read_sql_query(QUERY,conn)

#Plot the graph using the values.  Group 0 respondents are developers by profession, group 1 respondents are not.
df.NumRespondents.plot(kind='barh', stacked=False, figsize=(20, 10))

plt.title('Bar Chart of MainBranch')
plt.xlabel('Number of People')
plt.ylabel('Developer By Profession?')

plt.show()
```
